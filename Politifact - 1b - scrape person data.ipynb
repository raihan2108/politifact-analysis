# Scrape politifact's pages of who they are fact checking

```python
import numpy as np
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
```

# Load the list of URLs to scrape

```python
df = pd.read_csv('/gh/data/politifact/20190701scrape/raw_processed.csv', index_col=0)
urls_to_scrape = df['personalities_page'].unique()
```

# Get 1 html result to practice parsing

```python
# stopped at 105
seconds_pause = 2
n = len(urls_to_scrape)
for i_scrape, url in enumerate(urls_to_scrape):
    url_name = url.split('/')[2]
    try:
        # Get html from that page
        page_url = 'https://www.politifact.com{}'.format(url)
        response = requests.get(page_url)
        soup = BeautifulSoup(response.text, 'html.parser')
        pf_html = soup.prettify()

        # Save raw html
        file = open('/gh/data/politifact/20190707scrape_personalities/{}.html'.format(url_name),'w') 
        file.write(pf_html)
        file.close()

        # Pause
        print('{}/{}, scraped page {}'.format(i_scrape+1, n, url))
        time.sleep(seconds_pause)
    except Exception as e:
        print('\nfailed to scrape page {}\n{}\n'.format(url, e))
```
